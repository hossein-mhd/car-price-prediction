## Data preprocessing

```python
import datetime
```

```python
date_time = datetime.datetime.now()
```

```python
data['age'] = date_time.year - data['year']
```

```python
data.head()
```

```python
data.drop('year', axis=1, inplace=True)
```

```python
data.head()
```

```python
import seaborn as sns
```

```python
sns.boxplot(data['selling_price'])
```

```python
sorted(data['selling_price'], reverse=True)
```

```python
data[(data['selling_price'] >= 8150000)  & (data['selling_price'] <= 8900000)]
```

```python
data = data[~(data['selling_price'] >= 8150000)  & (data['selling_price'] <= 8900000)]

# removing 2 rows of the dataset
```

```python
data.shape
```

## Encoding the categorical columns

```python
data.head(1)
```

```python
data['fuel'].unique()
```

```python
data['fuel'] = data['fuel'].map({'Petrol':0, 'Diesel':1, 'CNG':2, 'LPG':3, 'Electric':4})
```

```python
data['fuel'].unique()
```

```python
data['seller_type'].unique()
```

```python
data['seller_type'] = data['seller_type'].map({'Dealer':0, 'Individual':1, 'Trustmark Dealer':2})
```

```python
data['seller_type'].unique()
```

```python
data['transmission'].unique()
```

```python
data['transmission'] = data['transmission'].map({'Manual':0, 'Automatic':1})
```

```python
data['transmission'].unique()
```

```python
data.head()
```

```python
data['owner'].unique()
```

```python
data['owner'] = data['owner'].map({'First Owner':0 ,'Second Owner':1 , 'Fourth & Above Owner':2 ,'Third Owner':3 , 'Test Drive Car':4})
```

```python
data['owner'].unique()
```

```python
data.head()
```

## Store feature matrix in x and response(target) in vector y

```python
x = data.drop(['name', 'selling_price'], axis=1)
y = data['selling_price']
```

```python
x
```

```python
y
```

## Spliting dataset into training set and test set

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

## Import the models

```python
data.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

## Model training

```python
lr = LinearRegression()
lr.fit(x_train, y_train)

rf = RandomForestRegressor()
rf.fit(x_train, y_train)

xgb = GradientBoostingRegressor()
xgb.fit(x_train, y_train)

xg = XGBRegressor()
xg.fit(x_train, y_train)
```

## Prediction on test data

```python
y_pred1 = lr.predict(x_test)
y_pred2 = rf.predict(x_test)
y_pred3 = xgb.predict(x_test)
y_pred4 = xg.predict(x_test)
```

## Evaluate the model

```python
from sklearn import metrics
```

```python
score1 = metrics.r2_score(y_test, y_pred1)
score2 = metrics.r2_score(y_test, y_pred2)
score3 = metrics.r2_score(y_test, y_pred3)
score4 = metrics.r2_score(y_test, y_pred4)
```

```python
print(score1, score2, score3, score4)
```

```python
final_data = pd.DataFrame({'Models':['LR','RF','GBR','XG'],
                          'R2_score':[score1, score2, score3, score4]})
```

```python
final_data
```

```python
sns.barplot(x='Models', y='R2_score', data=final_data)
```
